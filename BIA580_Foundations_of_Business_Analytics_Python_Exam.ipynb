```python
# Q1) 1
import numpy as np

def transpose(M):
  # Args: M: A 3x3 / Square matrix.
  # Returns the transpose of M.
  # The transpose of a square matrix creates another square matrix due to the same number of rows and columns.
  # We need to shuffle the items present in rows with the one present in columns
  result = [[M[j][i] for j in range(len(M))] for i in range(len(M[0]))]
  return result

def transpose_np(M):
  M=np.array(M)
  return M.T

def inverse(M):
  # Args: M: A 3x3 / Square matrix.
  # Returns the inverse of M.
  try:
    inverse_matrix = np.linalg.inv(M)
  except np.linalg.LinAlgError:
    # Exception that the input matrix is not invertible, return None.
    return "Matrix is not invertible"
  return inverse_matrix

def add_matrices(A,B):
    # The result of the list comprehension is a new list of lists,
    # where each inner list contains the sum of the corresponding elements in the A and B matrices for a given row.
    result = [[A[i][j] + B[i][j]  for j in range(len(A[0]))] for i in range(len(A))]
    return result

def add_matrices_np(A,B):
  # Using nuympy
    A = np.array(A)
    B = np.array(B)
    return np.add(A,B)

def subtract_matrices(A,B):
    # The result of the list comprehension is a new list of lists,
    # where each inner list contains the sum of the corresponding elements in the A and B matrices for a given row.
    result = [[A[i][j] - B[i][j]  for j in range(len(A[0]))] for i in range(len(A))]
    return result

def subtract_matrices_np(A,B):
    # Using nuympy
    A = np.array(A)
    B = np.array(B)
    return np.subtract(A,B)

def divide_matrices(A,B):
  AT = transpose_np(A)
  B_inv = inverse(B)
  result =  np.matmul(AT,B_inv)
  return result

# 6
A = [[1, 2, 3, 1],[0, 1, 0, 0],[4, 5, 6, 4],[0, 1, 0, 0]]
print('1) Transpose AT',transpose_np([[1, 2, 3], [4, 5, 6], [7, 8, 9]]))
print('2) Inverse B_inv',inverse([[1, 1, 1],[2, 3, 4],[4, 5, 5]]))
print('3) Added matrix C',add_matrices([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [[1, 2, 3], [4, 5, 6], [7, 8, 9]]))
print('4) Subtracted matrix D',subtract_matrices([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [[1, 2, 3], [4, 5, 6], [7, 8, 9]]))
print('5) Divide matrices',divide_matrices([[1,2,3],[4 ,5,6],[7 ,8,9]] ,[[5,8,1],[6,7,3],[4,5,9]]))
print('6) Transpose A : ',transpose_np(A))
print('6) Inverse A : ',inverse(A))
```

```python
# Q2)
import numpy as np

def transpose_find_row_max_sum(M):
  M = np.array(M)
  #Transpose the matrix first
  MT = M.T
  #Initialize the max sum to zero and an array/list to store row with the max_sum
  max_sum = 0
  rows_with_max_sum = []
  # iterate over the rows in the transposed array to find the sum of the row
  for row in MT:
    row_sum = sum(row)
    # if the row_sum is greater than max_sum, assign row_sum to max_sum and append row to the list
    if row_sum > max_sum:
      max_sum = row_sum
      rows_with_max_sum = [row]
    # This is to handle the corner case if the row_sum is equal to max_sum
    elif row_sum == max_sum:
      rows_with_max_sum.append(row)

  return rows_with_max_sum

print(transpose_find_row_max_sum([[1, 2, 3], [4, 5, 6], [7, 8, 9]]))
```

```python
# Q3)
import matplotlib.pyplot as plt
import numpy as np

def find_quadratic_roots(a, b, c):
    coefficients = [a, b, c]
    roots = np.roots(coefficients)
    return roots

def plot_quadratic_equation(a, b, c, roots):
  # Range of x-values to plot
  x = np.linspace(-5, 5, 400)
  # Calculate Y for the quadratic equation at respectove X.
  y = a * x**2 + b * x + c

  # Plot the quadratic equation.
  plt.plot(x, y)

  # Highlight the roots of the quadratic equation.
  for root in roots:
    plt.scatter(root, 0, color='red', marker='.', s=100)

  # Set the title and labels for the graph.
  plt.title('Quadratic Equation Plot')
  plt.xlabel('X')
  plt.ylabel('Y')

  # Show the graph.
  plt.grid(True)
  plt.show()

# Detailed Usage:
a = 3
b = 2
c = 6

# Plot the quadratic equation and highlight the roots.
plot_quadratic_equation(a, b, c, find_quadratic_roots(a, b, c))
```

```python
# Q5)
import numpy as np
import matplotlib.pyplot as plt

def find_intersection():
  # Define the coefficients of equations
  A = np.array([[3, 4], [1, -1]])
  b = np.array([120, 10])
  # Solve the system of equations
  solution = np.linalg.solve(A, b)
  # Extract the values of x and y from the solution
  x, y = solution

def coffee_sales():
  # Combined Earnings = Equation 1: 3x + 4y = 120  =>  y = (120 - 3x) / 4
  # Difference in Cups = Equation 2: x - y = 10   =>  y = x - 10
  x = np.linspace(0, 40, 400)  # Create an array of x
  y1 = (120 - 3 * x) / 4
  y2 = x - 10

  # Create a plot
  plt.plot(x, y1, label='3x + 4y = 120', color='blue')
  plt.plot(x, y2, label='x - y = 10', color='red')

  # Find the intersection point of the two lines (through find_intersection())
  intersection_x = 22.85  # Example value
  intersection_y = intersection_x - 10
  plt.plot(intersection_x, intersection_y, 'go', label='Solution (x, y)')

  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Graphical Representation of the System of Equations')
  plt.grid(True)
  plt.legend()
  plt.show()

coffee_sales()
```

```python
# Q6)
import math

def vector_magnitude(vector):
    # Calculate the magnitude of a 2D vector
    x, y = vector
    return math.sqrt(x**2 + y**2)

def vector_analysis(vectors):
    # Initialize variables to store the largest magnitude and corresponding vectors
    largest_magnitude = 0
    largest_magnitude_vectors = []

    for vector in vectors:
        magnitude = vector_magnitude(vector)
        # If the current magnitude is greater than the largest magnitude, update the variables
        if magnitude > largest_magnitude:
            largest_magnitude = magnitude
            largest_magnitude_vectors = [vector]
        # If the current magnitude is equal to the largest magnitude, add it to the list
        elif magnitude == largest_magnitude:
            largest_magnitude_vectors.append(vector)

    return largest_magnitude_vectors

# Example usage:
vectors = [(3, 4), (1, 2), (5, 12), (3, 4)]
result = vector_analysis(vectors)

print("Vectors with the largest magnitude:")
for vector in result:
    print(vector)
```

```python
# Q7)
import sympy as sp

def evaluate_limit(expression, variable, value):
  # Convert the expression and variable to sympy objects.
  expression = sp.sympify(expression)
  variable = sp.Symbol(variable)

  # Try to evaluate the limit using sympy.
  try:
    limit = sp.limit(expression, variable, value)
  except sp.PoleError:
    # If the limit is indeterminate, return Error.
    return 'Inderminate Limit Error'

  return limit

def plot_expression(expression, variable, value):
  # Define the symbol x, expression
  x = sp.symbols('x')
  expression = (x**2 - 4) / (x - 2)

  # Calculate the limit
  limit_value = sp.limit(expression, x, 2)

  x_values = np.linspace(-5, 5, 100)
  # Calculate the expression at the x
  y_values = [(x**2 - 4) / (x - 2) for x in x_values]

  # Plot the expression
  plt.plot(x_values, y_values, label='(x^2 - 4) / (x - 2)')

  # Highlight the limit point
  plt.plot(2, limit_value, 'ro', label=f'Limit at x=2: {limit_value}')

  # Set the title and labels for the graph.
  plt.title('Expression Plot')
  plt.xlabel(variable)
  plt.ylabel('Y')
  plt.grid(True)
  # Show the graph.
  plt.legend()
  plt.show()

#Usage:
expression = '(x**2 - 4) / (x-2)'
variable = 'x'
value = 2
# Print the limit.
print('The limit is:', evaluate_limit(expression, variable, value))
plot_expression(expression, variable, value)
```
