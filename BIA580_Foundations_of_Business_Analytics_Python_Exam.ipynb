```python
# Q2)
import numpy as np

def transpose_find_row_max_sum(M):
  M = np.array(M)
  #Transpose the matrix first
  MT = M.T
  #Initialize the max sum to zero and an array/list to store row with the max_sum
  max_sum = 0
  rows_with_max_sum = []
  # iterate over the rows in the transposed array to find the sum of the row
  for row in MT:
    row_sum = sum(row)
    # if the row_sum is greater than max_sum, assign row_sum to max_sum and append row to the list
    if row_sum > max_sum:
      max_sum = row_sum
      rows_with_max_sum = [row]
    # This is to handle the corner case if the row_sum is equal to max_sum
    elif row_sum == max_sum:
      rows_with_max_sum.append(row)

  return rows_with_max_sum

print(transpose_find_row_max_sum([[1, 2, 3], [4, 5, 6], [7, 8, 9]]))
```

```python
# Q4)

import sympy as sp
import matplotlib.pyplot as plt

def plot_polynomial_derivative(polynomial, derivative, r_start, r_end):
  # Convert the polynomials to Python functions
  f_polynomial = sp.lambdify(x, polynomial, 'numpy')
  f_derivative = sp.lambdify(x, derivative, 'numpy')

  # Generate x values for the plot
  x_values = np.linspace(r_start, r_end, 100)

  # Calculate y for the polynomial and its derivative
  y_polynomial = f_polynomial(x_values)
  y_derivative = f_derivative(x_values)

  # Plot the expression
  plt.plot(x_values, y_polynomial, label='Polynomial', linestyle='-', color='blue')
  # Plot the derivative
  plt.plot(x_values, y_derivative, label="Derivative", linestyle='--', color='red')
  # Set the title and labels for the graph.
  plt.title('Polynomial and Derivative Plot')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.grid(True)
  plt.legend()
  plt.show()

# Defining variable x
x = sp.symbols('x')
# Given Polynomial
polynomial = sp.exp(5*x) + (2/3)*x**5 - 1/sp.exp(7)
# Find the derivative of the expression
derivative = sp.diff(polynomial, x)

# Calling the function with the polynomial, derivative and range
plot_polynomial_derivative(polynomial,derivative, -1,1)
```

```python
# Q5)

# Price per cup for each shop
price_shop1 = 3
price_shop2 = 4
# Total earnings for the day
total_earnings = 120

def coffee_sales(x,y,total_earnings):
  # Number of cups sold by the first shop (x)
  # Number of cups sold by the second shop (y)
  # We will use a loop to search for the correct values of x and y
  for x in range(1, total_earnings // price_shop1 + 1):
      # Since the number of cups sold by the first shop is 10 more than the second shop
      y = x - 10

      # Calculate the total earnings from both shops
      total_earnings_shop1 = x * price_shop1
      total_earnings_shop2 = y * price_shop2
      total_earnings_combined = total_earnings_shop1 + total_earnings_shop2

      # Check if the total earnings match the given total_earnings
      if total_earnings_combined == total_earnings:
          print("Number of cups sold by the first shop:", x)
          print("Number of cups sold by the second shop:", y)
      else:
          print("No combination satisfied with integers")
          break

coffee_sales(3,4,120)
```

```python
# Q6)
import math

def vector_magnitude(vector):
    # Calculate the magnitude of a 2D vector
    x, y = vector
    return math.sqrt(x**2 + y**2)

def vector_analysis(vectors):
    # Initialize variables to store the largest magnitude and corresponding vectors
    largest_magnitude = 0
    largest_magnitude_vectors = []

    for vector in vectors:
        magnitude = vector_magnitude(vector)
        # If the current magnitude is greater than the largest magnitude, update the variables
        if magnitude > largest_magnitude:
            largest_magnitude = magnitude
            largest_magnitude_vectors = [vector]
        # If the current magnitude is equal to the largest magnitude, add it to the list
        elif magnitude == largest_magnitude:
            largest_magnitude_vectors.append(vector)

    return largest_magnitude_vectors

# Example usage:
vectors = [(3, 4), (1, 2), (5, 12), (3, 4)]
result = vector_analysis(vectors)

print("Vectors with the largest magnitude:")
for vector in result:
    print(vector)
```

```python
# Q7)
import sympy as sp

def evaluate_limit(expression, variable, value):
  # Convert the expression and variable to sympy objects.
  expression = sp.sympify(expression)
  variable = sp.Symbol(variable)
  # Try to evaluate the limit using sympy.
  try:
    limit = sp.limit(expression, variable, value)
  except sp.PoleError:
    # If the limit is indeterminate, return Error.
    return 'Inderminate Limit Error'

  return limit

def plot_expression(expression, variable, value):
  # Define the symbol x, expression
  x = sp.symbols('x')
  # Calculate the limit
  limit_value = sp.limit(expression, x, 2)

  x_values = np.linspace(-5, 5, 100)
  # Calculate the expression at the x
  y_values = [(x**2 - 4) / (x - 2) for x in x_values]

  # Plot the expression
  plt.plot(x_values, y_values, label='(x^2 - 4) / (x - 2)')

  # Highlight the limit point
  plt.plot(2, limit_value, 'ro', label=f'Limit at x=2: {limit_value}')

  # Set the title and labels for the graph.
  plt.title('Expression Plot')
  plt.xlabel(variable)
  plt.ylabel('Y')
  plt.grid(True)
  # Show the graph.
  plt.legend()
  plt.show()

#Calling the function:
expression = '(x**2 - 4) / (x-2)'
variable = 'x'
value = 2
# Print the limit.
print('The limit is:', evaluate_limit(expression, variable, value))
plot_expression(expression, variable, value)
```
